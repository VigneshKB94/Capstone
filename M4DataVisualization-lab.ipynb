# Hands-on Lab


## Visualizing distribution of data


```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = "SELECT ConvertedComp FROM master"
df = pd.read_sql_query(QUERY,conn)
df.dropna(subset=['ConvertedComp'], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df
count, bin_edges = np.histogram(df)

print(count)
print(bin_edges)

count, bin_edges = np.histogram(df)

print(count)
print(bin_edges)

df.plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.title('Histogram of Developers\' Annual Salary in USD')
plt.ylabel('Number of Respondents')
plt.xlabel('Annual Salary')

plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = "SELECT Age FROM master"
df = pd.read_sql_query(QUERY,conn)
df
df.plot(kind='box', figsize=(10,8))

plt.title('Box plot of Developers\' Age')
plt.ylabel('Age')

plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = "SELECT Age, WorkWeekHrs FROM master"
df = pd.read_sql_query(QUERY,conn)
df
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(11, 8), color='green')

plt.title('Developers\' Age and Weekly Working Hours')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

import plotly.express as px
QUERY = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master"
df = pd.read_sql_query(QUERY,conn)
df
df_new = df.dropna(axis=0)
fig = px.scatter(df_new, x=["WorkWeekHrs", "CodeRevHrs"], y="Age", size="Age", title="Code Review Hours and Weekly Work Hours vs Age", size_max=60)
fig.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = "SELECT Respondent, DatabaseWorkedWith, COUNT(DatabaseWorkedWith) as count FROM DatabaseWorkedWith group by Respondent"
df = pd.read_sql_query(QUERY,conn)
df[(df['DatabaseWorkedWith'] == 'MySQL') & (df['count'] == 1)]
QUERY = "SELECT Respondent, DatabaseWorkedWith, COUNT(DatabaseWorkedWith) as count FROM DatabaseWorkedWith group by DatabaseWorkedWith HAVING DatabaseWorkedWith LIKE '%SQL%' order by count desc"
df = pd.read_sql_query(QUERY,conn)
df
QUERY = "SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) as count FROM DatabaseDesireNextYear group by DatabaseDesireNextYear order by count desc LIMIT 5"
df = pd.read_sql_query(QUERY,conn)
df
fig = px.pie(df, values='count', names='DatabaseDesireNextYear', title='TOP 5 Databases Developers desire to learn next year')
fig.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = "SELECT WorkWeekHrs, CodeRevHrs, Age FROM master WHERE Age BETWEEN 30 AND 35"
df = pd.read_sql_query(QUERY,conn)
df_new = df.groupby(['Age'], as_index=False).median()
df_new.set_index('Age', inplace = True)
df_new
df_new.plot(kind='area',
             stacked=True,
             figsize=(15, 7))

plt.title('Median of Developers\' Weekly Work Hours and Code Review Hours in age group 30 to 35')
plt.ylabel('Hours')
plt.xlabel('Age')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = "SELECT ConvertedComp, Age FROM master WHERE Age BETWEEN 45 AND 60"
df = pd.read_sql_query(QUERY,conn)
df_cc = df.dropna(subset = ['ConvertedComp'], axis=0)
df_cc.reset_index(drop=True, inplace=True)
df_cc
df_cc = df_cc.groupby(['Age'], as_index=True).median()
df_cc.plot(kind="line", figsize = (14,8))
plt.title('Median of Developers\' Annual Salary for all ages from 45 to 60')
plt.ylabel('ConvertedComp')
plt.xlabel('Age')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = "SELECT MainBranch, COUNT(*) as Number_of_Respondents FROM master group by MainBranch order by Number_of_Respondents"
df = pd.read_sql_query(QUERY,conn)
df
fig = px.bar(df, x="Number_of_Respondents", y="MainBranch", orientation='h')
fig.show()
```
